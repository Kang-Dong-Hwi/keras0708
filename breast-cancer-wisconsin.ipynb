```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

TEST = 630
FILE_PATH = "C:/Projects/keras_talk/keras/breast-cancer-wisconsin.data"
dataset = np.loadtxt(FILE_PATH, delimiter=',' , dtype=np.int32 )

for idx in range(len(dataset)):
    res = dataset[idx][-1]
    dataset[idx][-1] = (res - 2) / 2
    
'''
#Number of Instances : 699

#Attribute Information:
   #  Attribute                     Domain
   -- -----------------------------------------
   1. Sample code number            id number
   2. Clump Thickness               1 - 10
   3. Uniformity of Cell Size       1 - 10
   4. Uniformity of Cell Shape      1 - 10
   5. Marginal Adhesion             1 - 10
   6. Single Epithelial Cell Size   1 - 10
   7. Bare Nuclei                   1 - 10
   8. Bland Chromatin               1 - 10
   9. Normal Nucleoli               1 - 10
  10. Mitoses                       1 - 10
  11. Class:                        (2 for benign, 4 for malignant)
  
  
#. Missing attribute values: 16

   There are 16 instances in Groups 1 to 6 that contain a single missing 
   (i.e., unavailable) attribute value, now denoted by "?". 
    
  
  
#Class distribution:
   Benign: 458 (65.5%)
   Malignant: 241 (34.5%)
'''


x_train = dataset[:TEST, 1:10]
y_train = dataset[:TEST, 10]

x_test = dataset[TEST:, 1:10]
y_test = dataset[TEST:, 10]



model = tf.keras.Sequential()
L1 = tf.keras.layers.Dense(20, input_dim=9, kernel_initializer='uniform', activation='relu')
L2 = tf.keras.layers.Dense(30, kernel_initializer='uniform', activation='relu')
L3 = tf.keras.layers.Dense(30, kernel_initializer='uniform', activation='relu')
L4 = tf.keras.layers.Dense(1, kernel_initializer='uniform', activation='sigmoid')

model.add(L1)
model.add(L2)
model.add(L3)
model.add(L4)

'''
from IPython.display import SVG
from tensorflow.keras.utils import model_to_dot

%matplotlib inline
SVG(model_to_dot(model, show_shapes=True, dpi=70).create(prog='dot', format='svg'))
'''

opt = Adam(lr=0.001, beta_1=0.9, beta_2=0.999)
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100, batch_size=63)

scores = model.evaluate(x_test, y_test)
print("{}: {}%".format(model.metrics_names[1], scores[1]*100))

```
